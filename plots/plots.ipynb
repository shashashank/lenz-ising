```python
from scipy.optimize import curve_fit
till = 100; simTime =100; calcSteps=10; trials = 10
x = np.arange(0, 100, 1/calcSteps)
def I_n(x, n):
    return jv(n, x)


# Parameters
J = 1.0
T = 1.5
# eta = np.tanh(J/T)

# alpha = 1.0
T0 = 10.0
# eta0 = np.tanh(J/T0)
# gamma = np.tanh(2*J/T)
# Define the function r(t)
def r(t, eta=np.tanh(J/T), alpha=1.0, eta0=np.tanh(J/T0), gamma=np.tanh(2*J/T)):
    # Exponential decay term
    exp_term = np.exp(-2 * alpha * t)
    # gamma =np.tanh(2*J/1.5)
    # Compute the sum
    total_sum = 0
    max_terms = 1000
    for l in range(1, max_terms + 1):
        term1 = (eta0**l - eta**l) * (I_n(2 * alpha * gamma * t, 1 - l) - I_n(2 * alpha * gamma * t, 1 + l))
        total_sum += term1
    return -(eta + exp_term * total_sum)
    

trials = 100
mag = np.zeros([simTime*calcSteps,trials])
energy = np.zeros([simTime*calcSteps,trials])
folder = '../data/glauber/19-12-2024/data'
for i in range(trials):
    time = np.loadtxt(folder+str(i), delimiter='\t')  # Change delimiter to ',' if comma-separated
    mag[:,i] = time[:,0]
    energy[:,i] = time[:,1]
y = np.mean(mag[:till,:], 1); std = np.std(mag[:till,:],1)
plt.plot(x[:till], y, linewidth=1, label='sim avg')
plt.fill_between(x[:till], y-std, y+std,  alpha=0.3)
plt.legend()
plt.xlabel('Time')
plt.ylabel('Magnetization')
plt.title('Magnetization vs. Time')
plt.grid(True)
plt.tight_layout()
# plt.savefig('magnetizationGlauber.pdf', dpi=300)
plt.show()


y = np.mean(energy[:till,:], 1); std = np.std(energy[:till,:],1)
# popt, pcov = curve_fit(r, x, y)
# eta, alpha, eta0, gamma
plt.hlines(-(1024-1)*np.tanh(J/T)/1024, 0,till/calcSteps, colors='orange', label='E (exact)')
plt.plot(x[:till], y, linewidth=1, label='sim avg;'+str(trials)+' trials')
plt.fill_between(x[:till], y-std, y+std,  alpha=0.3)
plt.plot(x[:till], r(x[:till]), 'k-.', linewidth=0.5, label=r'Glauber; $l=1000$')
plt.xlabel('Sweeps')
plt.ylabel('Energy')
plt.title(r'Energy vs. Time T=10$\to$T=1.5')
plt.legend()
plt.grid(True)
plt.tight_layout()
# plt.savefig('energyGlauber.pdf', dpi=300)
plt.show()
```
